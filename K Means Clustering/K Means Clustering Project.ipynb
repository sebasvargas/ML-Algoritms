## K Means Cluster.

```python
# Import Kmeans Cluster from sklean.
from sklearn.cluster import KMeans
# Create an instance of a Kmeans Model with 2 clusters (to private and public clasiffication).
kmeans = KMeans(n_clusters=2)
# First we need drop from dataset the column "Private", later w'll fit the model.
kmeans.fit(data.drop('Private',axis=1))
```

```python
# Show the cluster center vectors (Result of model).
print(kmeans.cluster_centers_)
# The cluster centers shape its one for each variable and one for each cluster. So shape of the centers it's 2x17.
print('Shape Cluster centers: '+str(kmeans.cluster_centers_.shape))
```

```python
# Labels of K-Means it's the "prediction" of the model and we will show.
print(kmeans.labels_)
print('Shape K-Means Labels: '+str(kmeans.labels_.shape))
```

### Evaluation
For clustering no exist a perfect metod for evaluate the clusters,  now with labels of the original dataset we use to evaluate. (In real life this no happen always).

```python
# We create a new column for df called 'Cluster', when 1 it's for private and 0 for public university.
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
data['Cluster'] = data['Private'].apply(converter)
data.head()
```

```python
# To evaluate the model, we will create a confusion matrix and classification report.
from sklearn.metrics import confusion_matrix,classification_report, ConfusionMatrixDisplay

print("\n"+"Classification Report: "+ "\n")
print(classification_report(data['Cluster'],kmeans.labels_))

print("Confusion Matrix: \n")
matrix = confusion_matrix(kmeans.labels_, data['Cluster'])
disp = ConfusionMatrixDisplay(confusion_matrix=matrix)
disp.plot()
plt.show()

```
